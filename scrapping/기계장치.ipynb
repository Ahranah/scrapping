````python

def extract_table_headers(driver, table_selector, max_attempts=3):
    """
    Selenium WebDriverë¥¼ ì‚¬ìš©í•˜ì—¬ í…Œì´ë¸”ì˜ í—¤ë”ë¥¼ ì¶”ì¶œí•©ë‹ˆë‹¤.

    Args:
        driver: Selenium WebDriver ê°ì²´.
        table_selector: í…Œì´ë¸”ì„ ì°¾ëŠ” CSS ì…€ë ‰í„°.
        nodata_class: 'nodata' í™•ì¸ì„ ìœ„í•œ í´ë˜ìŠ¤ ì´ë¦„. ê¸°ë³¸ê°’ì€ "td.nodata".
        max_attempts: í…Œì´ë¸” ë°ì´í„°ë¥¼ ì¶”ì¶œí•˜ê¸° ìœ„í•œ ì‹œë„ íšŸìˆ˜. ê¸°ë³¸ê°’ì€ 3.

    Returns:
        headers (list): í…Œì´ë¸” í—¤ë”(ë‚ ì§œ ë“±)ë¥¼ í¬í•¨í•œ ë¦¬ìŠ¤íŠ¸.
    """
    try:
        # í…Œì´ë¸” ë°ì´í„° ì¶”ì¶œ ì‹œë„
        for attempt in range(max_attempts):
            try:
                # í…Œì´ë¸” ìš”ì†Œ ì°¾ê¸°
                table_element = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.CSS_SELECTOR, table_selector))
                )
                print("í…Œì´ë¸” ìš”ì†Œ ë°œê²¬")

                # í…Œì´ë¸” í—¤ë”(ë‚ ì§œ ë“±) ì¶”ì¶œ
                header_row = table_element.find_elements(By.CSS_SELECTOR, "thead tr th span")
                headers = [header.text.strip() for header in header_row if header.text.strip()]
                print("Headers:", headers)
                return headers  # ì„±ê³µ ì‹œ í—¤ë” ë°˜í™˜
            except Exception as e:
                print(f"Attempt {attempt + 1}/{max_attempts} failed: {e}")
                time.sleep(1)  # ì¬ì‹œë„ ì „ ëŒ€ê¸°

        print("í…Œì´ë¸” ë°ì´í„°ë¥¼ ì—¬ëŸ¬ ë²ˆ ì‹œë„í–ˆì§€ë§Œ ì‹¤íŒ¨í–ˆìŠµë‹ˆë‹¤.")
        return None  # ëª¨ë“  ì‹œë„ ì‹¤íŒ¨ ì‹œ None ë°˜í™˜

    except Exception as e:
        print(f"í…Œì´ë¸” ì¶”ì¶œ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
        return None
    
def check_nodata(driver, nodata_class="td.nodata", nodata_message="ì¡°íšŒëœ ìë£Œê°€ ì—†ìŠµë‹ˆë‹¤."):
    """
    Selenium WebDriverë¥¼ ì‚¬ìš©í•˜ì—¬ íŠ¹ì • í´ë˜ìŠ¤ì˜ ìš”ì†Œì—ì„œ 'nodata' ë©”ì‹œì§€ë¥¼ í™•ì¸í•˜ëŠ” í•¨ìˆ˜.

    Args:
        driver: Selenium WebDriver ê°ì²´.
        nodata_class (str): 'nodata' í…ìŠ¤íŠ¸ë¥¼ í™•ì¸í•˜ê¸° ìœ„í•œ í´ë˜ìŠ¤ ì´ë¦„. ê¸°ë³¸ê°’ì€ "td.nodata".
        nodata_message (str): 'nodata' ë©”ì‹œì§€ë¡œ ê°„ì£¼í•  í…ìŠ¤íŠ¸. ê¸°ë³¸ê°’ì€ "ì¡°íšŒëœ ìë£Œê°€ ì—†ìŠµë‹ˆë‹¤."

    Returns:
        bool: 'nodata' ë©”ì‹œì§€ê°€ ìˆìœ¼ë©´ True, ì—†ìœ¼ë©´ False.
    """
    try:
        # ì§€ì •ëœ í´ë˜ìŠ¤ ì´ë¦„ì˜ ìš”ì†Œë“¤ ì°¾ê¸°
        no_data_elements = driver.find_elements(By.CLASS_NAME, nodata_class)
        
        # ê° ìš”ì†Œì˜ í…ìŠ¤íŠ¸ì—ì„œ nodata_message í™•ì¸
        if no_data_elements and any(nodata_message in el.text for el in no_data_elements):
            print("nodata ë©”ì‹œì§€ê°€ ë°œê²¬ë˜ì—ˆìŠµë‹ˆë‹¤.")
            return True  # nodata ë©”ì‹œì§€ê°€ ìˆìŒ
        else:
            print("nodata ë©”ì‹œì§€ê°€ ì—†ìŠµë‹ˆë‹¤.")
            return False  # nodata ë©”ì‹œì§€ê°€ ì—†ìŒ

    except Exception as e:
        print(f"nodata í™•ì¸ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
        return False
    
def extract_row(driver, depth = "depth-1", wait_time=10):
    """
    í…Œì´ë¸”ì—ì„œ ê°’ì— í•´ë‹¹í•˜ëŠ” í–‰ ('depth-i')ì„ ì¶”ì¶œí•˜ëŠ” í•¨ìˆ˜.
    
    Args:
        driver: Selenium WebDriver ê°ì²´.
        wait_time (int): ëŒ€ê¸° ì‹œê°„ (ì´ˆ). ê¸°ë³¸ê°’ì€ 10ì´ˆ.
    
    Returns:
        list: ì¶”ì¶œëœ ì²« ë²ˆì§¸ í–‰ ë°ì´í„°. (ì˜ˆ: ['ë§¤ì¶œì•¡(*)', '10,180,665', ...])
    """
    try:
        print("ğŸ” í…Œì´ë¸” ëŒ€ê¸° ì¤‘...")
        # í…Œì´ë¸” ìš”ì†Œê°€ ë¡œë“œë  ë•Œê¹Œì§€ ëŒ€ê¸°
        table_element = WebDriverWait(driver, wait_time).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "div.finance-statement table.details"))
        )
        print("âœ… í…Œì´ë¸” ë¡œë“œ ì™„ë£Œ")

        # ì²« ë²ˆì§¸ í–‰ (depth-1) ì°¾ê¸°
        first_row = WebDriverWait(table_element, wait_time).until(
            lambda table: table.find_element(By.CSS_SELECTOR, f"tr.{depth}")
        )
        print("âœ… ì²« ë²ˆì§¸ í–‰ ì°¾ê¸° ì™„ë£Œ")

        # 'td span' ìš”ì†Œì—ì„œ í…ìŠ¤íŠ¸ ì¶”ì¶œ
        spans = first_row.find_elements(By.CSS_SELECTOR, "td span")
        row_data = [span.text.strip() for span in spans if span.text.strip()]
        print(f"âœ… ì¶”ì¶œëœ ë°ì´í„°: {row_data}")
        return row_data

    except Exception as e:
        print(f"âŒ ì²« ë²ˆì§¸ í–‰ ì¶”ì¶œ ì‹¤íŒ¨: {e}")
        return []
````
